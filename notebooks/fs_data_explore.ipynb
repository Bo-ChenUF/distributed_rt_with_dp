In this notebook, we do some analysis on the distributed reproduction numbers of french school example.

```python
# import packages
import numpy as np
import pandas as pd
import sys
import os
sys.path.append(os.path.join(os.path.dirname("__file__"), ".."))
import matplotlib.pyplot as plt
```

```python
# load data
df = pd.read_csv("../data/reproduction_number_matrix.csv", header=None)
print(f"The reproduction number has {df.shape[0]} rows and {df.shape[1]} columns.")
df.head()
```

```python
# convert to numpy array
matrix = df.to_numpy()
matrix_nonzero = matrix[np.nonzero(matrix)]

print(f"The matrix have overall {matrix.reshape(-1).shape[0]} entries!")
print(f"The matrix have overall {matrix_nonzero.shape[0]} non-zero entries!")
print(f"The rate of non-zero entries: {matrix[np.nonzero(matrix)].shape[0] / matrix.reshape(-1).shape[0]}.")
```

```python
# more analysis
print(f"The value of minimal nonzero entry is {np.min(matrix_nonzero)}")
print(f"The value of maximal nonzero entry is {np.max(matrix_nonzero)}")
```
